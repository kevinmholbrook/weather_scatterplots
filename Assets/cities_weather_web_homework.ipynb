```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import time
```

```python
#Resources/cities.csv
cities_data = pd.read_csv("../Resources/cities.csv")
```

```python
#Convert into a DataFrame
cities_data_pd = pd.DataFrame(cities_data)
cities_data_pd.head()
```

```python
#Export cities_data_pd into a CSV File
cities_data_pd.to_csv("../Resources/cities.csv", encoding="UTF-8", index=False)
```

```python
#Define Metrics
lats = cities_data_pd["Lat"]
max_temps = cities_data_pd["Max Temp"]
humidity = cities_data_pd["Humidity"]
cloudiness = cities_data_pd["Cloudiness"]
wind_speed = cities_data_pd["Wind Speed"]
```

```python
#Build Scatterplot for Latitude vs Temperature 
plt.scatter(lats,max_temps, edgecolor="black", color="violet",marker="o",alpha=.8,label="Cities")
plt.title("City Latitude vs. Max Temperature (%s)" % time.strftime("%x"))
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

#Save the scatterplot
plt.savefig("../Resources/Plot_1.png")
```

```python
#Build Scatterplt for Latitude vs Humidity 
plt.scatter(lats,humidity, edgecolor="black", color="green",marker="o",alpha=.8,label="Cities")
plt.title("City Latitude vs. Humidity (%s)" % time.strftime("%x"))
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)

#Save the scatterplot
plt.savefig("../Resources/Plot_2.png")
```

```python
#Build Scatterplt for Latitude vs Cloudiness 
plt.scatter(lats,cloudiness, edgecolor="black", color="red",marker="o",alpha=.8,label="Cities")
plt.title("City Latitude vs. Cloudiness (%s)" % time.strftime("%x"))
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)

#Save the scatterplot
plt.savefig("../Resources/Plot_3.png")
```

```python
#Build Scatterplt for Latitude vs Wind_Speed 
plt.scatter(lats,wind_speed, edgecolor="black", color="orange",marker="o",alpha=.8,label="Cities")
plt.title("City Latitude vs. Wind Speed (%s)" % time.strftime("%x"))
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)

#Save the scatterplot
plt.savefig("../Resources/Plot_4.png")
```
